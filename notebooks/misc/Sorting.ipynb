```python
def insertion_sort(arr):
    idx = 0
    for idx in range(len(arr)):
        print('idx: {}'.format(idx))
        jdx = idx
        # swap successive pairs
        while jdx > 0 and (arr[jdx-1] > arr[jdx]):
            print('jdx: {}'.format(jdx))
            arr[jdx-1], arr[jdx] = arr[jdx], arr[jdx-1]
            print(arr)
            jdx -= 1
    return arr
                    
insertion_sort([8,7,6,5,4,3,2,1])
```

```python
def merge(a, b):
    # a, b are sorted
    out = []
    while a and b: 
        if a[0] < b[0]:
            out.append(a.pop(0))
        else:
            out.append(b.pop(0))
    # one or the other is guaranteed to be empty after this
    out += a
    out += b
    return out
```

```python
out = merge([1,3,7], [2,5,9])
print(out)
out = merge([], [1])
print(out)
out = merge([1], [])
print(out)
out = merge([2, 5], [6, 7])
print(out)
```

```python
def merge_pointers(a, b):
    # a, b are sorted
    i_a, i_b = 0, 0
    len_a, len_b = len(a), len(b)
    out = []
    while (i_a < len_a) and (i_b < len_b):
        if a[i_a] < b[i_b]:
            out.append(a[i_a])
            i_a += 1
        else:
            out.append(b[i_b])
            i_b += 1
    out += a[i_a:]
    out += b[i_b:]
    return out
```

```python
out = merge_pointers([1,3,7], [2,5,9])
print(out)
out = merge_pointers([], [1])
print(out)
out = merge_pointers([1], [])
print(out)
out = merge_pointers([2, 5], [6, 7])
print(out)
```

```python
def merge_sort(arr):
    # recursively split 
    half = len(arr)//2
    if len(arr) <= 1:
        return arr
    lhs = merge_sort(arr[:half])
    rhs = merge_sort(arr[half:])
    print('LHS: {}, RHS: {}'.format(lhs, rhs))
    return merge(lhs, rhs)
    
        
```

```python
# The original partition scheme described by C.A.R. Hoare uses two indices that start at the ends of the array being partitioned, 
# then move toward each other, until they detect an inversion: a pair of elements, one greater than or equal to the pivot, 
# one lesser or equal, that are in the wrong order relative to each other. The inverted elements are then swapped.
# When the indices meet, the algorithm stops and returns the final index. 

def partition(arr, lo, hi):
    pivot = arr[(lo + hi)//2]
    left = lo
    right = hi
    print('start: arr[left]: {}, arr[right]: {}, pivot: {}'.format(
        arr[left], arr[right], pivot))
    while left <= right:
        print('before: {}'.format((left, right)))
        while arr[left] < pivot:
            left += 1
            print('left: {}, arr[left]: {}'.format(left, arr[left]))
        while arr[right] > pivot:
            print('right: {}, arr[right]: {}'.format(right, arr[right]))
            right -= 1
        print('after: {}'.format((left, right)))
        if (left <= right):
            print('swapping left/right: {}, {}'.format(arr[left], arr[right]))
            arr[left], arr[right] = arr[right], arr[left]
            left += 1
            right -= 1
    return left

def partition2(arr):
    N = len(arr)
    pivot = arr[(N-1)//2]
    left = 0
    right = N-1
    while left <= right:
        print('before: {}'.format((left, right)))
        while arr[left] < pivot:
            left += 1
        while arr[right] > pivot:
            right -= 1
        print('after: {}'.format((left, right)))
        if (left <= right):
            arr[left], arr[right] = arr[right], arr[left]
            left += 1
            right -= 1
    print('pivot item: {} index: {}'.format(pivot, left))
    print('exiting: {}'.format(arr))
    return left
```

```python
inp = ['e', 'a', 'd', 'b', 'c']
pivot = partition(inp, 0, 4)
print(inp)
print(pivot)

inp = ['e', 'a', 'd', 'b', 'c']
pivot = partition2(inp)
print(inp)
print(pivot)
```

```python
def partition(ll, start, stop):
    if start == stop:
        return
    pindex = (start+stop) // 2
    pivot = ll[pindex]
    lo = start
    hi = stop
    while True:
        while lo <= hi and ll[hi] >= pivot:
            print(f"decrement lo={lo}, hi={hi}, pivot={pivot}, ll[lo]={ll[lo]}, ll[hi]={ll[hi]}")
            hi -= 1
        while lo <= hi and ll[lo] <= pivot:
            print(f"increment lo={lo}, hi={hi}, pivot={pivot}, ll[lo]={ll[lo]}, ll[hi]={ll[hi]}")
            lo += 1
        if lo <= hi:
            print(f"lo <= hi: lo={lo}, hi={hi}, ll={ll}, pindex={pindex}")
            ll[lo], ll[hi] = ll[hi], ll[lo]
            print(f"swapped: lo={lo}, hi={hi}, ll={ll}, pindex={pindex}")
        else:
            print(f"lo > hi: lo={lo}, hi={hi}, ll={ll}, pindex={pindex}")
            break
    print(f"crossed: lo={lo}, hi={hi}, ll={ll}")
    if pindex == lo:
        return pindex
    elif pindex > lo:
        ll[pindex], ll[lo] = ll[lo], ll[pindex]
        print(f"crossed, swapped pindex and lo: lo={lo}, hi={hi}, ll={ll}, pindex={pindex}")
        return lo
    else:
        ll[pindex], ll[hi] = ll[hi], ll[pindex]
        print(f"crossed, swapped pindex and hi: lo={lo}, hi={hi}, ll={ll}, pindex={pindex}")
    return hi
```

```python
def quicksort(ll, start=None, end=None):
    if start is None:
        start = 0
    # can't use "if not end"
    # recursive call means end could be 0
    if end is None:
        end = len(ll) - 1
    if start >= end:
        return
    p = partition(ll, start, end)
    quicksort(ll, start, p-1)
    quicksort(ll, p+1, end)
```

```python
ll = [1,5,6,8,3]
quicksort(ll)
```
