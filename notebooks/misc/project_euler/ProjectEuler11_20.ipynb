What is the value of the first triangle number to have over five hundred divisors?

```python
def factorise(n):
    if n <= 0:
        raise ValueError("Only works for positive n: got {}".format(n))
    if n == 1:
        return n
    factors = []
    factor = 2
    # get rid of 2s
    # in separate loop because 
    # in main loop we have to increment by 2,
    # here only increment by 1. 
    while n %2 == 0:
        factors.append(2)
        n = n//2

    # try all odd numbers.
    factor = 3
    while factor <= n:
        #print(n)
        while n%factor == 0:
            factors.append(factor)
            n = n // factor
        factor += 2
    return factors

def powerset(seq):
    # use binary count. 
    n = len(seq)
    retval = []
    for i in range(0, 2**n):
        thisval = []
        # work out what binary maps to this number
        bitstring = '{:b}'.format(i).zfill(n)
        for idx, bit in enumerate(bitstring):
            if int(bit):
               thisval.append(seq[idx])
        if thisval:
            retval.append(thisval)
    return retval

def divisors(n):
    # find prime factors
    # take all set combinations of those prime factors
    pfactors = factorise(n)
    combinations = powerset(pfactors)
    #print(combinations)
    factors = [reduce(mul, x) for x in combinations]
    factors = [1] + factors + [n]
    return set(factors)

def triangle_number_stop(ndivs=5):
    # return the first triangle number with greater than ndivs divisors
    n = 1
    total = 1
    divs = {1}
    while len(divs) < ndivs:
        n += 1
        total += n
        divs = divisors(total)
        print(total, divs)
    return total
    
    
```

The following iterative sequence is defined for the set of positive integers:

n → n/2 (n is even)
n → 3n + 1 (n is odd)

Using the rule above and starting with 13, we generate the following sequence:

13 → 40 → 20 → 10 → 5 → 16 → 8 → 4 → 2 → 1
It can be seen that this sequence (starting at 13 and finishing at 1) contains 10 terms. Although it has not been proved yet (Collatz Problem), it is thought that all starting numbers finish at 1.

Which starting number, under one million, produces the longest chain?

NOTE: Once the chain starts the terms are allowed to go above one million.



```python
last = 1000000
# keys: start digit, values: path length
lengths = {1: 0}
maxn = 1

for n in range(2, last+1):
    current = n
    steps = 0
    while True:
        # if it reduces to a previous n, add the current steps
        # and stop
        if current in lengths:
            lengths[n] = steps + lengths[current]
            break
        if current == 1:
            lengths[n] = steps
            break
        elif current % 2 == 0:
            current = current/2
        else:
            current = current*3 + 1
        steps += 1
    if lengths[n] > lengths[maxn]:
        maxn = n
        
```

$2^{15} = 32768$ and the sum of its digits is 3 + 2 + 7 + 6 + 8 = 26.

What is the sum of the digits of the number $2^{1000}$?



```python
for n in range(1, 20):
    s = 2**n
    sumdigits = sum(map(int, str(s)))
    print('{}: {}'.format(s, sumdigits))
    
```

```python
def add_strings(digits):
    # digits: list of strings describing the digits
    # value is returned as a string. 
    total = []
    overflow = 0
    linelen = len(digits[0])
    # count backwards, assume all have same 

    for i in range(1, linelen+1):
        current = 0
        idx = -i
        for line in digits:
            current += int(line[idx])
        current += overflow
        current = str(current)
        total.insert(0, current[-1])
        if len(current) > 1:
            overflow = int(current[:-1])
        else:
            overflow = 0

    if overflow:
        total.insert(0, str(overflow))

    total = ''.join(total)
    return total

def powerof2(n):
    if n == 1:
        return '1'
    # do this using the addition routine with strings developed earlier. 
    digits = ['1', '1']
    for i in range(1, n+1):
        out = add_strings(digits)
        digits = [out, out]
    return out
```

If the numbers 1 to 5 are written out in words: one, two, three, four, five, then there are 3 + 3 + 5 + 4 + 4 = 19 letters used in total.

If all the numbers from 1 to 1000 (one thousand) inclusive were written out in words, how many letters would be used?


NOTE: Do not count spaces or hyphens. For example, 342 (three hundred and forty-two) contains 23 letters and 115 (one hundred and fifteen) contains 20 letters. The use of "and" when writing out numbers is in compliance with British usage.

```python
dictionary = {
    1: 'one',
    2: 'two',
    3: 'three',
    4: 'four',
    5: 'five',
    6: 'six',
    7: 'seven',
    8: 'eight',
    9: 'nine',
    10: 'ten',
    11: 'eleven',
    12: 'twelve',
    13: 'thirteen',
    14: 'fourteen',
    15: 'fifteen',
    16: 'sixteen',
    17: 'seventeen',
    18: 'eighteen',
    19: 'nineteen',
    20: 'twenty',
    30: 'thirty',
    40: 'forty',
    50: 'fifty',
    60: 'sixty',
    70: 'seventy',
    80: 'eighty',
    90: 'ninety',
    100: 'hundred',
    1000: 'thousand',
}

def decompose(n):
    return {
        1000: n//1000,
        100: (n%1000)//100,
        10: (n%100)//10,
        1: n%10,
    }

def format(n):
    factors = decompose(n)
    #print(factors)
    retval = []
    ones = factors[1]
    tens = factors[10]
    hundreds = factors[100]
    thousands = factors[1000]
    # combination of ones and tens determines whether we will use words
    ones_tens = ones + tens*10
    if ones_tens and ones_tens <= 20:
        retval.append(dictionary[ones_tens])
    else:
        if ones:
            if not tens:
                retval.append(dictionary[ones])
            else:
                multiple = tens*10
                retval.append('{}-{}'.format(dictionary[multiple], dictionary[ones]))
        else:
            if tens:
                multiple = tens*10
                retval.append(dictionary[multiple])
    if hundreds:
        # we only want to add "and" if there is already something. 
        if retval:
            retval.insert(0, 'and')
        retval.insert(0, '{} hundred'.format(dictionary[hundreds]))
    if thousands:
        if retval:
            retval.insert(0, 'and')
        retval.insert(0, '{} thousand'.format(dictionary[thousands]))
    return " ".join(retval)
```

You are given the following information, but you may prefer to do some research for yourself.

* 1 Jan 1900 was a Monday.
* Thirty days has September, April, June and November.
* All the rest have thirty-one,
* Saving February alone, Which has twenty-eight, rain or shine.
* And on leap years, twenty-nine.
* A leap year occurs on any year evenly divisible by 4, but not on a century unless it is divisible by 400.
* How many Sundays fell on the first of the month during the twentieth century (1 Jan 1901 to 31 Dec 2000)?



```python
# Let Monday be encoded as 0 because it's easy
# algorithm:
# assign each day of each year a cardinal number
# for each year, calculate the day of week of every first day of month
# keep track of the Sundays. 

def is_leap_year(n):
    if n%100 == 0:
        if n%400 == 0:
            return True
        else:
            return False
    elif n%4 == 0:
        return True
    return False

def days_in_month(y, m):
    if m == 2:
        if is_leap_year(y):
            return 29
        else:
            return 28
    elif m in (1, 3, 5, 7, 8, 10, 12):
          return 31
    return 30
    
def weekday_firstdayof_month(y, m):
    # number of years between n and 1900
    between = y-1900
    days = sum([(366 if is_leap_year(i) else 365) for i in range(1900, y)])
    # add on the months
    days += sum([days_in_month(y, i) for i in range(1, m)])
    # first day of 1900 was Monday which is weekday 0
    mod = days%7
    # "mod" gives the number of weekdays offset from Monday
    return mod

collect = {}
for year in range(1900, 2001):
    for month in range(1, 13):
        wd = weekday_firstdayof_month(year, month)
        if wd == 6:
            collect[(year, month)] = True
```

n! means n × (n − 1) × ... × 3 × 2 × 1

For example, 10! = 10 × 9 × ... × 3 × 2 × 1 = 3628800,

and the sum of the digits in the number 10! is 3 + 6 + 2 + 8 + 8 + 0 + 0 = 27.

Find the sum of the digits in the number 100!



```python
def sumofdigits_factorial(n):
    # find the sum of digits in n!
    # throw away any trailing zeros and keep multiplying.
    total = 1
    # current sum of digits
    current = 0
    i = 0
    while i < n:
        i += 1

        total *= i
        if total%10 == 0:
            while total %10 == 0:
                total //=10
        current = sum(map(int, str(total)))
    return total, current
    
```
