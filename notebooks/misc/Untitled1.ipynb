```python
import matplotlib 
%matplotlib inline

from matplotlib import pyplot as plt
matplotlib.rcParams['figure.figsize'] = [15, 12]

import numpy as np
```

```python
def logistic(x):
    ex = np.exp(x)
    return ex/(1+ex)

def logit(x):
    return np.log(x/(1-x))
```

```python
x = np.linspace(-5, 5, 1000)

ye = logistic(x)
yl = logit(x)

plt.plot(x, ye)
plt.plot(x, yl)

plt.grid()
```
